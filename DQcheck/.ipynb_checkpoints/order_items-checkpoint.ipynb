```python
import pandas as pd
```

```python
order_items=pd.read_csv('../Landing/order_items/order_items.csv')
```

```python
order_items.shape
```

```python
order_items.isnull().sum()
```

```python
order_items['item_id'] = order_items['item_id'].apply(pd.to_numeric, errors='coerce')
```

```python
order_items=order_items.dropna(subset=['item_id'])
```

```python
order_items.tail()
```

```python
duplicate_count = order_items.duplicated().sum()
print(duplicate_count)
```

```python
duplicate_count = order_items.duplicated(keep=False)

duplicate_count_df=order_items[duplicate_count]
```

```python
sorted_duplicate_df = duplicate_count_df.sort_values(by=['order_id', 'item_id'], ascending=True)
```

```python
sorted_duplicate_df
```

```python
order_items = order_items.drop_duplicates()
```

```python
order_items.info()
```

```python
# Filter rows where the price is within the range of 200 to 10000
order_items = order_items[(order_items['list_price'] >= 200) & (order_items['list_price'] <= 10000)]
```

```python
products=pd.read_csv('../Landing/products/products.csv')
```

```python
products.head()
```

```python
# Extract the 'item_id' column from the 'products' DataFrame
product_item_ids = products['product_id']

# Filter order_items to include only the items that exist in the 'products' DataFrame
order_items = order_items[order_items['item_id'].isin(product_item_ids)]

```

```python
order_items.head()
```

```python
is_available = order_items['item_id'].isin(product_item_ids)
```

```python
import os

# Define the parent directory path
parent_directory = '../staging_1'

# Define the subdirectory path
subdirectory = 'order_items'

# Create the parent directory if it doesn't exist
if not os.path.exists(parent_directory):
    os.makedirs(parent_directory)
    print(f"Parent directory '{parent_directory}' created.")

# Create the subdirectory if it doesn't exist
if not os.path.exists(os.path.join(parent_directory, subdirectory)):
    os.makedirs(os.path.join(parent_directory, subdirectory))
    print(f"Subdirectory '{subdirectory}' created.")

# Define the file path where you want to save the CSV file
file_path = os.path.join(parent_directory, subdirectory, 'order_items.csv')

# Save the DataFrame as a CSV file
order_items.to_csv(file_path, index=False)

print(f"CSV file saved successfully in the '{subdirectory}' subfolder within the '{parent_directory}' folder.")
```
